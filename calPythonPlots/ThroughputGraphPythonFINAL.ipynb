```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\Khushbu\\Downloads\\tp1.csv');
df
```

```python
df200_20 = df.query('Distance == 208 and Width == 20')
df200_50 = df.query('Distance == 208 and Width == 50')
df300_20 = df.query('Distance == 300 and Width == 20')
df300_50 = df.query('Distance == 300 and Width == 50')
df250_20 = df.query('Distance == 250 and Width == 20')
df250_50 = df.query('Distance == 250 and Width == 50')

df200_20
```

```python
mean200_20 = df200_20["Throughput"].mean() #ide = 3.510962	
mean200_50 = df200_50["Throughput"].mean() #3.510961919
mean300_20 = df300_20["Throughput"].mean()  #3.754887502
mean300_50 = df300_50["Throughput"].mean() #ide = 3.459432
mean250_20 = df250_20["Throughput"].mean() #3.510962	
mean250_50 = df250_50["Throughput"].mean()

ideMean_200_20 = df200_20["IndexDif"].mean()
ideMean_200_50 = df200_50["IndexDif"].mean()
ideMean_300_20 = df300_20["IndexDif"].mean()
ideMean_300_50 = df300_50["IndexDif"].mean()
ideMean_250_20 = df250_20["IndexDif"].mean()
ideMean_250_50 = df250_50["IndexDif"].mean()

x = pd.DataFrame()
y = pd.DataFrame()

y['Throughput'] = [mean200_20, mean200_50, mean300_20, mean300_50, mean250_20, mean250_50]

x['IndexOfD'] = [ideMean_200_20, ideMean_200_50, ideMean_300_20, ideMean_300_50, ideMean_250_20, ideMean_250_50]
x
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

plt.xlabel('Mean Index of Difficulty', fontsize=16)
plt.ylabel('Mean Throughput', fontsize=16)


regr = linear_model.LinearRegression()
regr.fit(x, y);

y_pred = regr.predict(x);

plt.scatter(x, y)
plt.plot(x, y_pred, linewidth=3)
plt.show()
```
